```python
import pandas as pd

bace_ki = pd.read_csv("bace_ki_unscaled.csv")
bace_ki.describe()
```

### Log transformation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Perform log transformation
ki_values_log = np.log(bace_ki)
ki_values_log.to_csv('bace_ki_values_log_transformation.csv', index=False)

# Plot the original and transformed distributions
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(bace_ki, bins=20, edgecolor='black')
plt.title('Original Ki Values')
plt.xlabel('Ki Values')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(ki_values_log, bins=20, edgecolor='black')
plt.title('Log-Transformed Ki Values')
plt.xlabel('Log(Ki) Values')
plt.ylabel('Frequency')



plt.tight_layout()
plt.show()
```

```python
# Kernel Density Plot
sns.kdeplot(ki_values_log, shade=True)
plt.title("Kernel Density Plot Log-Transformed Ki Values")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```

```python
ki_values_log.describe()
```

```python
from scipy.stats import shapiro
import pandas as pd

def shapiro_test(data):
    stat, p = shapiro(data)
    print(stat, p)

    if p > 0.05:
        print(f'data is likely Gaussian (p={p:.3f})')
    else:
        print(f'data is not likely Gaussian (p={p:.3f})')
```

```python
shapiro_test(ki_values_log)
```

### Square Root Transformation

```python
# Perform square root transformation
ki_values_sqrt = np.sqrt(bace_ki)

# Plot the original and transformed distributions
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(bace_ki, bins=20, edgecolor='black')
plt.title('Original Ki Values')
plt.xlabel('Ki Values')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(ki_values_sqrt, bins=20, edgecolor='black')
plt.title('Square Root-Transformed Ki Values')
plt.xlabel('sqrt(Ki) Values')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
shapiro_test(ki_values_sqrt)
```
