```python
def get_fingerprint_name (path_str):
    # Path to the XML file
    xml_file_path = path_str

    # Find the indices of "_xml\\" and ".xml" in the path
    start_index = xml_file_path.find("fingerprint_output\\descriptor_output_") + len("fingerprint_output\\descriptor_output_")
    end_index = xml_file_path.rfind(".csv")

    if start_index != -1 and end_index != -1:
        extracted_string = xml_file_path[start_index:end_index]
        return extracted_string
    else:
        return "Pattern not found in the path"
```

```python
from lazypredict.Supervised import LazyRegressor
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
```

### using lazy predict

```python
for i in range(0,12):
    
    fingerprint_descriptor = pd.read_csv(padel_fingerprints_xml_output[i])
    fingerprint_descriptor_name = get_fingerprint_name (padel_fingerprints_xml_output[i])
    
    print(fingerprint_descriptor_name)
    
    X = fingerprint_descriptor.drop('Name', axis=1)
    X_train, X_test, y_train, y_test = train_test_split(X, bace_ki ,test_size=0.2, random_state = 117)

    #Initialize and run LazyRegressor
    reg = LazyRegressor(verbose=0, ignore_warnings=True, custom_metric=None)
    models, predictions = reg.fit(X_train, X_test, y_train, y_test)

    # Print the list of models and their performance metrics
    models.to_csv(f"bace_lazy_predict_ki_{fingerprint_descriptor_name}_model.csv")
    
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
import math

# get bace scaled ki 
bace_ki = pd.read_csv('bace_ki_minmax_scale.csv')
bace_ki.describe()

X = pd.read_csv('fingerprint_output/descriptor_output_ExtendedFingerprinter.csv')

X_train, X_test, y_train, y_test = train_test_split(X, bace_ki ,test_size=0.2, random_state = 117)
    
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
r2score = r2_score(y_test, y_predict)
mse = mean_squared_error(y_predict, y_test)
rmse = math.sqrt(mse)

print(f"r2score => {r2score} \n rmse => {rmse}" )
```
